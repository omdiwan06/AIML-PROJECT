# Cleaning

```python
backup = car.copy()
```

```python
#if year is string and numeric then keep it otherwise discard it
car = car[car['year'].str.isnumeric()]
```

```python
#convert year from object to integer

car['year'].astype(int)
```

```python
car['year'] = car['year'].astype(int)
```

```python
car['year'].unique()
```

```python
car=car[car['Price']!='Ask For Price']
```

```python
car['Price'] = car['Price'].str.replace(',','').astype(int)
```

```python
car['kms_driven'] = car['kms_driven'].str.split(' ').str.get(0).str.replace(',','')
```

```python
# As few km have entry as petrol 
car = car[car['kms_driven'].str.isnumeric()]
```

```python
car = car[~car['fuel_type'].isna()]
```

```python
car['name'] = backup['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
car['name'] = car['name'].str.lower()
```

```python
car
```

```python
#to reset the index

car.reset_index()
```

```python
#to remove extra column of old index
car = car.reset_index(drop =True)
```

```python
car['company'] = car['company'].str.lower()
```

```python
car.describe()
```

```python
#to find outlier in price
car[car['Price']>6e6]
```

```python
car = car[car['Price']<6e6].reset_index(drop = True)
```

# Model

```python
X = car.drop(['Price'], axis=1)
y = car['Price']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state= 433)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
```

```python
ohe.fit(X[['name','company','fuel_type']])
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories = ohe.categories_),['name','company','fuel_type']),remainder = 'passthrough')
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```
